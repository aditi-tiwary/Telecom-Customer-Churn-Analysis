### Data Cleaning

1. Create a copy of base data for manupulation & processing

```python
data = data.copy()
```

2. Converting Total Charges to numerical data type

```python
data.TotalCharges = pd.to_numeric(data.TotalCharges, errors='coerce')
data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column. Let's check these records

```python
data.loc[data ['TotalCharges'].isnull() == True]
```

4. Missing Value Treatementng.

```python
result = (11 / 7043) * 100
print(result)
```

Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
#Removing missing values 
data.dropna(how = 'any', inplace = True)
```

5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
# Get the max tenure
print(data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

data['tenure_group'] = pd.cut(data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
data['tenure_group'].value_counts()
```

6. Remove columns not required for processing

```python
#drop column customerID and tenure
data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
data.head()
```

### Data Exploration

1. Plot distibution of individual predictors by churn

## Univariate Analysis

```python
for i, predictor in enumerate(data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=data, x=predictor, hue='Churn')
    plt.grid(False)
    plt.show()
```

2. Convert the target variable 'Churn' in a binary numeric variable i.e. Yes = 1 ; No = 0

```python
data['Churn'] = np.where(data.Churn == 'Yes',1,0)
data.head()
```

3. Convert all the categorical variables into dummy variables

```python
data_dummies = pd.get_dummies(data)
data_dummies.head()
```

9. Relationship between Monthly Charges and Total Charges

```python
sns.lmplot(data=data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

As expected, Total Charges increases as Monthly Charges increases

10. Churn by Monthly Charges and Total Charges

```python
Mth = sns.kdeplot(data_dummies.MonthlyCharges[(data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Mth = sns.kdeplot(data_dummies.MonthlyCharges[(data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", fill = True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly Charges by Churn')
```

**Insight: Churn is high when Monthly Charges are high**

```python
Tot = sns.kdeplot(data_dummies.TotalCharges[(data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Tot = sns.kdeplot(data_dummies.TotalCharges[(data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", fill = True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total Charges by Churn')
```

Surprisingly, Churn is Higher at Lower Total Charges

However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total Charges then the picture ia s bit cle::- Higher Monthly Charsge aL loweT tenure results intL lower Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, LoweT tenure and Lower Total Charge are lienkd to High Churn.
